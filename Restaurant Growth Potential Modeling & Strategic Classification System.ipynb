# OUTLIER IDENTIFICATION

```python
num_cols = raw_df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    Q1 = raw_df[col].quantile(0.25)
    Q3 = raw_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = raw_df[(raw_df[col] < lower_limit) | (raw_df[col] > upper_limit)]
    print(f"\n{col}: {len(outliers)} outliers")

    # plt.figure(figsize=(8, 2))
    # plt.boxplot(raw_df[col], vert=False)
    # plt.title(f'Boxplot of {col}')
    # plt.xlabel(col)
    # plt.show()

```

# OUTLIER TREATMENT

```python
num_cols = raw_df.select_dtypes(include=['int64', 'float64']).columns

# Capping outliers using mask()
for col in num_cols:
    Q1 = raw_df[col].quantile(0.25)
    Q3 = raw_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    raw_df[col] = raw_df[col].mask(raw_df[col] < lower_limit, lower_limit)
    raw_df[col] = raw_df[col].mask(raw_df[col] > upper_limit, upper_limit)
```

# RESULTS AFTER TREATMENT

```python
for col in num_cols:
    Q1 = raw_df[col].quantile(0.25)
    Q3 = raw_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = raw_df[(raw_df[col] < lower_limit) | (raw_df[col] > upper_limit)]
    print(f"\n{col}: {len(outliers)} outliers")
```

# ENCODING (ONEHOTENCODER)

```python
from sklearn.preprocessing import OneHotEncoder

cat_cols = raw_df.select_dtypes(include=["object"]).columns

encoder = OneHotEncoder(drop="first", sparse_output=False)
encoded = encoder.fit_transform(raw_df[cat_cols])

encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(cat_cols), index=raw_df.index)
```

# SCALING 


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_num = scaler.fit_transform(raw_df[num_cols])

scaled_num_df = pd.DataFrame(scaled_num, columns=num_cols, index=raw_df.index)
```

# PCA DIMENSIONALITY REDUCTION

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.85, random_state=42)
X_pca = pca.fit_transform(df)

print("Original features:", df.shape[1])
print("PCA components:", X_pca.shape[1])
```

```python
X_scaled = scaled_num_df
```

# K - MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia_scores = []
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    inertia_scores.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, labels))

    print(f"K = {k}, Inertia = {kmeans.inertia_:.2f}, Silhouette = {silhouette_score(X_scaled, labels):.3f}")

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(range(2,11), inertia_scores, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")

plt.subplot(1,2,2)
plt.plot(range(2,11), silhouette_scores, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Analysis")

plt.show()
```

```python
final_k = 3

kmeans_model = KMeans(n_clusters=final_k, random_state=42)
df["KMeans_Cluster"] = kmeans_model.fit_predict(X_scaled)
```

```python
df["KMeans_Cluster"] = kmeans_model.fit_predict(X_scaled)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca_model = PCA(n_components=2, random_state=42)
X_pca_2d = pca_model.fit_transform(X_scaled)

plt.figure(figsize=(8, 5))
plt.scatter(
    X_pca_2d[:, 0],
    X_pca_2d[:, 1],
    c=df["KMeans_Cluster"],
    cmap="viridis",
    s=60,
    alpha=0.7
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("K-Means Clusters (PCA Projection)")
plt.colorbar(label="Cluster Label")
plt.tight_layout()
plt.show()

```

# CLUSTER LABELING

```python
cluster_labels = {
    0: "High-Growth, High-Risk Expanders",
    1: "Stable Core Performers",
    2: "Channel-Driven Growth Candidates"
}

df["Cluster_Label"] = df["Cluster"].map(cluster_labels)
```

# GPI Feature Selection

```python
gpi_features = [
    "GrowthFactor",
    "MonthlyOrders",
    "AOV",
    "COGSRate",
    "OPEXRate",
    "CommissionRate",
    "DeliveryRadiusKM"
]
```

# GPI Scaling

```python
from sklearn.preprocessing import StandardScaler

gpi_scaler = StandardScaler()
gpi_scaled = gpi_scaler.fit_transform(df[gpi_features])

gpi_df = pd.DataFrame(
    gpi_scaled,
    columns=gpi_features,
    index=df.index
)
```

# GPI Calculation

```python
df["GPI"] = (
    0.30 * gpi_df["GrowthFactor"] +
    0.25 * gpi_df["MonthlyOrders"] +
    0.15 * gpi_df["AOV"] -
    0.10 * gpi_df["COGSRate"] -
    0.10 * gpi_df["OPEXRate"] -
    0.05 * gpi_df["CommissionRate"] +
    0.05 * gpi_df["DeliveryRadiusKM"]
)
```

# Rank by Growth Potential

```python
df["GPI_Rank"] = df["GPI"].rank(method="dense", ascending=False)
```

# CLUSTER-WISE GPI SUMMARY

```python
cluster_gpi_summary = (
    df.groupby("Cluster_Label")[["GPI"]]
    .agg(["mean", "min", "max", "count"])
    .round(3)
)

cluster_gpi_summary
```

```python
df["GPI_Rank"] = df["GPI"].rank(ascending=False)
```

```python
df["GPI_Rank"].sort_values(ascending=True).head(5)
```

# 📊 Data Driven Resturant Classification (Orders, Revenue, Cuisine & GPI)

1️⃣ Aggregator dominance is structurally high
Uber Eats contributes the largest share of total orders (~800K) and highest revenue (~18K), confirming strong demand but also heavy platform dependency risk.

2️⃣ Revenue does not scale proportionally with orders across channels
Although Uber Eats leads both orders and revenue, Self-Delivery and DoorDash generate comparable revenue with significantly fewer orders, indicating better unit economics per order.

3️⃣ High-volume cuisines are not always high-growth cuisines
Pizza and Burgers lead in monthly orders, but only Pizza shows positive GPI, proving that demand alone does not imply growth readiness.

4️⃣ Cuisine-level monetization differs sharply from demand
Burgers and Indian cuisine generate high average order volumes, yet their GPI is weak or negative, suggesting cost pressure or channel imbalance limiting scalable growth.

5️⃣ Only a few cuisines are growth-ready
Pizza and Burgers are the only cuisines with positive GPI, signaling scalable expansion potential, while Thai, Kebabs/Mediterranean, and Indian show negative GPI → expansion risk.

6️⃣ Strategic insight for expansion
Growth strategy should prioritize high-GPI cuisines (Pizza, Burgers) while restructuring cost and channel mix for high-volume but low-GPI cuisines to avoid overextension.

#  Growth Levers Impacting Expansion Readiness

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))

# ---------- 1. DELIVERY LEVER ----------
plt.subplot(2, 2, 1)
delivery_ratio = final_df.assign(
    Aggregator_Ratio = (
        final_df["UberEatsOrders"] + final_df["DoorDashOrders"]
    ) / final_df["MonthlyOrders"]
).groupby("Cluster_Label")["Aggregator_Ratio"].mean()

bars1 = plt.bar(delivery_ratio.index, delivery_ratio.values, color="steelblue")
plt.title("Aggregator Dependency by Cluster")
plt.bar_label(bars1, fmt="%.2f")


# ---------- 2. DEMAND LEVER ----------
plt.subplot(2, 2, 2)
orders = final_df.groupby("Cluster_Label")["MonthlyOrders"].mean()

bars2 = plt.bar(orders.index, orders.values, color="green")
plt.title("Average Monthly Orders")
plt.bar_label(bars2, fmt="%.0f")


# ---------- 3. VALUE LEVER ----------
plt.subplot(2, 2, 3)
aov = final_df.groupby("Cluster_Label")["AOV"].mean()

bars3 = plt.bar(aov.index, aov.values, color="purple")
plt.title("Average Order Value (AOV)")
plt.bar_label(bars3, fmt="%.0f")


# ---------- 4. COST LEVER ----------
plt.subplot(2, 2, 4)
costs = final_df.groupby("Cluster_Label")[["COGSRate", "OPEXRate"]].mean()

bars4a = plt.bar(costs.index, costs["COGSRate"], label="COGS")
bars4b = plt.bar(costs.index, costs["OPEXRate"],
                 bottom=costs["COGSRate"], label="OPEX")

plt.title("Cost Pressure by Cluster")
plt.legend()
plt.bar_label(bars4a, fmt="%.2f")
plt.bar_label(bars4b, fmt="%.2f")

plt.tight_layout()
plt.show()
```

# Growth Levers Impacting Expansion Readiness

1️⃣ High-Growth restaurants handle ~55% of orders via aggregators, the lowest dependency, showing better channel control for expansion.

2️⃣ High-Growth cluster leads demand with ~1,531 monthly orders, nearly 90% higher than Stable Performers (804).

3️⃣ Average Order Value is flat (₹38–₹39) across all clusters, contributing <3% variation, so AOV is not a growth lever.
